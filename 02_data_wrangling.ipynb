# 1. Preparing Mexico Data

```python
import pandas as pd
```

```python
#Import

df1 = pd.read_csv("data/mexico-real-estate-1.csv")
df2 = pd.read_csv("data/mexico-real-estate-2.csv")
df3 = pd.read_csv("data/mexico-real-estate-3.csv")

# Print object type and shape for DataFrames
print("df1 type:", type(df1))
print("df1 shape:", df1.shape)
print()
print("df2 type:", type(df2))
print("df2 shape:", df2.shape)
print()
print("df3 type:", type(df3))
print("df3 shape:", df3.shape)
```

```python
#Insspecting our dataframe
df1.info()
df2.info()
df3.info()
```

# Clean df1

```python
df1.shape
```

```python
df1.info()
```

```python
df1.head()
```

```python
#Remove NaN values
#inplace=True says that it drop the values in the current dataset not in a new one
df1.dropna(inplace=True)
#df1.info()

#change the strings to floats
df1["price_usd"] = (
    df1["price_usd"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "")
    .astype(float)
)
```

```python
df1.info()
```

# Clean df2

```python
df2.shape
```

```python
df2.info()
```

```python
df2.head()
```

```python
#Remove NaN values (rows)
df2.dropna(inplace=True)

#Create a "price_usd" column
df2["price_mxn"] = (df2["price_mxn"] / 19).round(2)

#Drop "price_mxn" column
df2.drop(columns = ["price_mxn"], inplace=True)

df2.head()
```

# Clean df3

```python
df3.shape
```

```python
df3.info()
```

```python
df3.head()
```

```python
#df3["lat-lon"].head()
#drop "Nan" rows
df3.dropna(inplace=True)


#Split "lat-lon"
df3[["lat", "lon"]] = df3["lat-lon"].str.split(",", expand=True)

df3.head()
```

```python
#Create "state" col from "place_with_parent_names"
df3["state"] = df3["place_with_parent_names"].str.split("|", expand = True)[2]

#drop columns
df3.drop(columns = ["place_with_parent_names", "lat-lon"], inplace=True)
df3.head()
```

# Concatenate DataFrames


```python
df = pd.concat([df1, df2, df3])
print(df.shape)
df.head()
```

# Save df

```python
df.to_csv("data/mexico-real-estate-clean.csv", index=False)
```
